```python
import numpy as np

import pandas as pd

import altair as alt


import bokeh
import bokeh.plotting
from bokeh.plotting import ColumnDataSource
from bokeh.models import LabelSet
import bokeh.io
bokeh.io.output_notebook()
```

```python
df = pd.read_excel('073019_Deuterium_Transfer_Peak_Areas.xlsx')
df2 = pd.read_excel('073019_Deuterium_Transfer_Peak_Areas.xlsx')
```

```python
df.tail()
```

```python
nonanalAreas = np.repeat(df['Area'].loc[df['Peak ID'] == 'Nonanal'],3).reset_index().drop("index", axis=1)
df['Normalized Area'] = np.divide(df['Area'],nonanalAreas['Area'])
df.head()
```

```python
df_plotting = df.loc[df['Peak ID'] != 'Nonanal']

alt.Chart(df_plotting
    ).mark_point(
    ).encode(
        x='Species:N',
        y=alt.Y('Normalized Area',
                type='quantitative',
                scale=alt.Scale(type='log')),
        color='Peak ID:N')
```

```python
df_C24 = df.loc[df['Peak ID'] == 'D50C24']

p = bokeh.plotting.figure(plot_width=1200,
                          plot_height=600,
                          title='C24',
                          x_axis_label='Species',
                          y_axis_label='D50C24/nonanal',
                          y_axis_type='log')


for i in range(20):
    curSample = df_C24.loc[df_C24['WellNumber'] == i+1]
    if (any(curSample['Species'] == 'S')):
        curLiometopum=curSample['Normalized Area'].loc[curSample['Species']=='L']
        curSceptobius=curSample['Normalized Area'].loc[curSample['Species']=='S']
        
        p.circle(1,curLiometopum,color='Blue',legend='Liometopum',line_width=4)
        p.circle(2,curSceptobius,color='Red',legend='Sceptobius',line_width=4)
        p.line([1,2],[curLiometopum,curSceptobius],color='black')
        
        source = ColumnDataSource(
            pd.DataFrame(
                dict(
                    normalized_area=[curLiometopum.values],
                    position=[1],
                    labels=[str(i+1)]
                )
            )
        )
        
        label=bokeh.models.LabelSet(x='position',y='normalized_area',text='labels',level='glyph',
                                   x_offset=-25, y_offset=-10, source=source, render_mode='canvas')
        p.add_layout(label)
        
    if (any(curSample['Species'] == 'P')):
        curLiometopum=curSample['Normalized Area'].loc[curSample['Species']=='L']
        curPlatyusa=curSample['Normalized Area'].loc[curSample['Species']=='P']
        
        p.circle(3,curLiometopum,color='Blue',legend='Liometopum',line_width=4)
        p.circle(4,curPlatyusa,color='Green',legend='Platyusa',line_width=4)
        p.line([3,4],[curLiometopum,curPlatyusa],color='black')
        
        source = ColumnDataSource(
            pd.DataFrame(
                dict(
                    normalized_area=[curLiometopum.values],
                    position=[3],
                    labels=[str(i+1)]
                )
            )
        )
        
        label=bokeh.models.LabelSet(x='position',y='normalized_area',text='labels',level='glyph',
                                   x_offset=-25, y_offset=-10, source=source, render_mode='canvas')
        p.add_layout(label)
        
    if (any(curSample['Species'] == 'D')):
        curLiometopum=curSample['Normalized Area'].loc[curSample['Species']=='L']
        curDalotia=curSample['Normalized Area'].loc[curSample['Species']=='D']
        
        p.circle(5,curLiometopum,color='Blue',legend='Liometopum',line_width=4)
        p.circle(6,curDalotia,color='Purple',legend='Dalotia',line_width=4)
        p.line([5,6],[curLiometopum,curDalotia],color='black')
        
        source = ColumnDataSource(
            pd.DataFrame(
                dict(
                    normalized_area=[curLiometopum.values],
                    position=[5],
                    labels=[str(i+1)]
                )
            )
        )
        
        label=bokeh.models.LabelSet(x='position',y='normalized_area',text='labels',level='glyph',
                                   x_offset=-25, y_offset=-10, source=source, render_mode='canvas')
        p.add_layout(label)
        
    if (any(curSample['Species'] == 'A')):
        curLiometopum=curSample['Normalized Area'].loc[curSample['Species']=='L']
        curAnchoma=curSample['Normalized Area'].loc[curSample['Species']=='A']
        
        p.circle(7,curLiometopum,color='Blue',legend='Liometopum',line_width=4)
        p.circle(8,curAnchoma,color='Orange',legend='Anchoma',line_width=4)
        p.line([7,8],[curLiometopum,curAnchoma],color='black')
        
        source = ColumnDataSource(
            pd.DataFrame(
                dict(
                    normalized_area=[curLiometopum.values],
                    position=[7],
                    labels=[str(i+1)]
                )
            )
        )
        
        label=bokeh.models.LabelSet(x='position',y='normalized_area',text='labels',level='glyph',
                                   x_offset=-25, y_offset=-10, source=source, render_mode='canvas')
        p.add_layout(label)
        
    if (any(curSample['Species'] == 'Blank')):
        curLiometopum=curSample['Normalized Area'].loc[curSample['Species']=='L']
        curBlank=curSample['Normalized Area'].loc[curSample['Species']=='Blank']
        
        p.circle(9,curLiometopum,color='Blue',legend='Liometopum',line_width=4)
        p.circle(10,curBlank,color='Cyan',legend='Blank',line_width=4)
        p.line([9,10],[curLiometopum,curBlank],color='black')
        
        source = ColumnDataSource(
            pd.DataFrame(
                dict(
                    normalized_area=[curLiometopum.values],
                    position=[9],
                    labels=[str(i+1)]
                )
            )
        )
        
        label=bokeh.models.LabelSet(x='position',y='normalized_area',text='labels',level='glyph',
                                   x_offset=-25, y_offset=-10, source=source, render_mode='canvas')
        p.add_layout(label)
        
p.xaxis.visible = False
p.xgrid.visible = False
bokeh.io.show(p)
```
