```python
import numpy as np

import skimage
import skimage.io
import pandas as pd
import bokeh
import bokeh.plotting
import bokeh.io
from bokeh.models import ColumnDataSource
from bokeh.io import export_svgs


bokeh.io.output_notebook()
```

```python
im = skimage.io.imread('./190505_DCHC_Transfer_SP_02_well_7/img59232.png')[:,:,2]
ylen,xlen=im.shape
```

```python
df = pd.read_csv('./190505_DCHC_Transfer_SP_02_well_7/CollectedData_Tom.csv',header=[1,2])
df.head()
```

```python
headx=df['anthead']['x'][10]
heady=df['anthead']['y'][10]
gasterx=df['antgaster']['x'][10]
gastery=df['antgaster']['y'][10]

ant_length=np.sqrt((headx-gasterx)**2 + (heady-gastery)**2)
t = np.linspace(0, 2*np.pi, 200)
xc = (headx+gasterx)/2
yc = (heady+gastery)/2
aval = 0.7*ant_length
bval = np.sqrt(aval**2-(ant_length/2)**2)
phi=np.arctan((heady-gastery)/(headx-gasterx))


xplot = xc + aval*np.cos(t)*np.cos(phi) - bval*np.sin(t)*np.sin(phi)
yplot = yc + aval*np.cos(t)*np.sin(phi) + bval*np.sin(t)*np.cos(phi)
```

```python
p = bokeh.plotting.figure(x_range=(0,ylen), y_range=(0,xlen))

p.image(image=[im], x=0, y=0, dw=xlen, dh=ylen, palette="Viridis256")
p.circle(df['anthead']['x'][10],df['anthead']['y'][10],size=10,color='red')
p.circle(df['antgaster']['x'][10],df['antgaster']['y'][10],size=10,color='red')
p.circle(df['beetlehead']['x'][10],df['beetlehead']['y'][10],size=10,color='orange')


p.line([df['anthead']['x'][10],df['antgaster']['x'][10]],
       [df['anthead']['y'][10],df['antgaster']['y'][10]],
       color='red',line_width=2)

p.line(xplot,yplot,color='red',line_width=2)
bokeh.io.show(p)
```
